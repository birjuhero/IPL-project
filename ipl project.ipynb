# player of the match top10 and the bar graph 

```python

man_of_theMatch=df_match['player_of_match'].value_counts()[0:10]
```

```python
len(man_of_theMatch)
```

```python
man_of_theMatchlist=list(df_match['player_of_match'].value_counts()[0:10].keys())
```

```python
plt.figure(figsize=(15,5))
plt.bar(man_of_theMatchlist,man_of_theMatch,color='r')
plt.show()
```

```python
plt.figure(figsize=(50,5))
plt.bar(list(df_match['player_of_match'].value_counts()[:].keys()),df_match['player_of_match'].value_counts()[:],color='r')
plt.xticks(rotation=30, horizontalalignment="center")#this is used to Rotate the x-labels by 30 degrees, and keep the text aligned horizontally
plt.show()
```
